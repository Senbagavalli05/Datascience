```python
import pandas as pd

# Sample data
data = {
    'Hotel ID': ['Hotel 1', 'Hotel 2', 'Hotel 3', 'Hotel 4', 'Hotel 5'],
    'Room Category': ['Category A', 'Category B', 'Category A', 'Category C', 'Category B'],
    'N': [1, 0, 1, 0, 1],
    'P': [2, 2, 3, 2, 4]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Define a function to create tour packages
def create_tour_packages(df):
    tour_packages = []

    for name, group in df.groupby(['P', 'N']):
        nights, room_primary = name
        if room_primary == 1:
            package_name = f'{nights} Nights Package'
            package_details = ', '.join(group['Hotel ID'] + ' - ' + group['Room Category'])
            tour_packages.append((package_name, package_details))

    tour_packages_df = pd.DataFrame(tour_packages, columns=['Package Name', 'Package Details'])

    return tour_packages_df

# Call the function to create tour packages
tour_packages_df = create_tour_packages(df)

# Display the resulting tour packages
print(tour_packages_df)
```
